```python
import pandas as pd
```

```python
payroll_data = pd.read_csv("../data/Citywide_Payroll_Data_(Fiscal_Year)_20251030.csv")
```

```python
payroll_data.head()
```

```python
payroll_data.dtypes
```

```python
payroll_data["Regular Gross Paid"] = payroll_data["Regular Gross Paid"].str.replace("$", " ", regex=False)
payroll_data["Total OT Paid"] = payroll_data["Total OT Paid"].str.replace('$', " ", regex=False)
```

```python
payroll_data["Regular Gross Paid"] = pd.to_numeric(payroll_data["Regular Gross Paid"], errors='coerce')
payroll_data["Total OT Paid"] = pd.to_numeric(payroll_data["Total OT Paid"], errors='coerce')
payroll_data["Base Salary"] = pd.to_numeric(payroll_data["Base Salary"], errors='coerce')
```

```python
payroll_data["Fiscal Year"] = payroll_data["Fiscal Year"].astype("category")
payroll_data["Title Description"] = payroll_data["Title Description"].astype("category")
payroll_data["Fiscal Year"] = payroll_data["Fiscal Year"].astype(object)
```

```python
payroll_data.dtypes
```

```python
doc_data = payroll_data[payroll_data["Agency Name"].apply(lambda x: "DEPARTMENT OF CORR" in str(x))]
```

```python
doc_data.head()
```

```python
doc_corrections_data_reduced = doc_data[doc_data["Regular Gross Paid"] ==  "per Annum"]

print(f"Number of rows for the original dataframe is \
{len(payroll_data)} and the number of rows for the reduced dataframe is {len(doc_data)}")
```

```python
doc_sum = doc_corrections_data_reduced.groupby("Fiscal Year")["Regular Gross Paid"].sum()
print(doc_sum)
```

```python
doc_corrections_data_reduced.groupby(["Title Description"]).agg(
                      {"Total OT Paid":"median",
                      "Base Salary":"median",
                      "Total OT Paid":"count"
                      }
      ).reset_index(
    
    ).sort_values(
        by="Base Salary",
)
```
